```python
def get_all_sessions(cohort_instance):
    """
    Gather all session dictionaries from the cohort and create Session objects.
    """
    sessions = []
    for mouse in cohort_instance.cohort["mice"]:
        for session_id in cohort_instance.cohort["mice"][mouse]["sessions"]:
            session_dict = cohort_instance.get_session(session_id)
            if session_dict:
                session = Session(session_dict)
                sessions.append(session)
    return sessions

sessions = get_all_sessions(cohort)
```
